# Gaussian Process Classifier

aussian Process Classifiers (GPC) are a different kind of machine learning algorithm for categorizing things. Imagine you have a bunch of data points with labels, like separating emails into spam or not-spam. A GPC builds a kind of map that predicts how likely a new data point is to belong to a certain category, based on how similar it is to the labeled data points you've already shown it. It's like learning from examples to classify new things, while also giving you an idea of how confident it is in those classifications.


Regression
$$
\hat{y}_* = \mathbf{k}(x_*, \mathbf{X}) (K + \sigma^2 \mathbf{I})^{-1} \mathbf{Y}
$$


Binary Classification
$$
p(y_* = 1 | x_*) = \sigma(f(x_*))
$$
