# Decision Tree Tutorial

In this project I will be analyzing the breast cancer dataset from `sklearn.datasets`. What I will be predicting is to see if a tumor is either benign or malignant. I will also be using `pandas` and `plotly.express` for EDA and data visualization.

```python
import pandas as pd
import plotly.express as px
from sklearn.datasets import load_breast_cancer
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
# Converting Everything into a DataFrame

data =  load_breast_cancer()
df = pd.DataFrame(data=data.data, columns = data.feature_names)
df['target'] = data.target
```

```python
df.head(10)
```

### Splitting the dataset into training and testing categories.

```python
# Feature Selection
X = df.copy()
y = df['target']

# Ensuring that both the feaure and the label are in df form and series form
if isinstance(X, pd.Series):
    X = X.to_frame()
    
if isinstance(y, pd.Series):
    y = pd.Series(y)

# Splitting the Data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
# Making the model

d_tree = DecisionTreeClassifier(criterion='gini')
d_tree.fit(X_train, y_train)
```

```python
d_tree.get_params()
```

```python
# Predicting the model
# With decsion trees there are multiple ways to make predictions
# This will give an array 0 being benign and 1 being malignant.

predictions = d_tree.predict(X_test)
print(predictions)
```

```python
# Classification Report

print(classification_report(y_test, predictions, target_names = ['malignant', 'benign']))
```
